```python
# 파일별로 내용 읽기

def get_conetents(file_list): # 파일 경로  + 이름이 있는 file_list
    y_class = []
    X_text = []
    class_dict = {
        1: "0", 2: "0", 3:"0", 4:"0", 5:"1", 6:"1", 7:"1", 8:"1"}

    for file_name in file_list: # 파일 경로 + 파일 제목을 1개씩 file_name에 넣어줌
        try:
            f = open(file_name, "r",  encoding="cp949") # cp949는 window 파일 코딩 타입
            category = int(file_name.split(os.sep)[1].split("_")[0]) #os.sep은 / or \ 같은 구분자.
            # ex: news_data\\8_Who Should Be Tottenham Hotspur's No. 1 Summer Transfer Target.txt 에서 sep으로 분리하면
            #[news_data, 8_Who Should Be Tottenham Hotspur's No. 1 Summer Transfer Target.txt]임.
            # 여기서 [1]을 _로 나누면  8
            y_class.append(class_dict[category]) # y_class에 "1"이 하나 올라감
            X_text.append(f.read()) # X_test에 Summer Transfer.. 파일의 내용만 올라감
            f.close() # 닫기
        except UnicodeDecodeError as e: # 에러 발생시 유니코드 에러로 취급
            print(e) # 유니코드 에러라고 알림
            print(file_name) # 에러 뜬 파일 경로 + 이름 보여줌
    return X_text, y_class # 이 값들 반환

X_text, y_class   = get_conetents(file_list) # == y_class, X_text   = get_conetents(file_list) 이거랑 다름, 조심
```

```python
def get_count_vector(text, corpus): 
    text = [sentence.split() for sentence in text] #텍스트는 띄어쓰기로 나눠주고
    # 단어단위로 끊어서 첫 번째 파일의 첫 번째 단어가 corpus에서 3509번째 인덱스값이라는 거
    word_number_list = [[corpus[get_cleaned_text(word)] for word in words] for words in text]
    # 4042개의 0이 들어간 1차원 매트릭스를 80개 만들어줌
    X_vector = [[0 for _ in range(len(corpus))] for x in range(len(text))]
    for i, text in enumerate(word_number_list):
        for word_number in text:
            X_vector[i][word_number] += 1
    return X_vector
X_vector = get_count_vector(X_text, corpus_dict); X_vector
```

```python
# gcs로 비교
import math
def get_cosine_similarity(v1,v2):
    "compute cosine similarity of v1 to v2: (v1 dot v2)/{||v1||*||v2||)"
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(v1)):
        x = v1[i]; y = v2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    return sumxy/math.sqrt(sumxx*sumyy)
```

```python
def get_similarity_score(X_vector, source):
    source_vector = X_vector[source]
    similarity_list = []
    for target_vector in X_vector:
        similarity_list.append(
            get_cosine_similarity(source_vector, target_vector))
    return similarity_list
```

```python
def get_top_n_similarity_news(similarity_score, n):
    import operator
    x = {i:v for i, v in enumerate(similarity_score)}
    sorted_x = sorted(x.items(), key=operator.itemgetter(1))

    return list(reversed(sorted_x))[1:n+1]

def get_accuracy(similarity_list, y_class, source_news):
    source_class = y_class[source_news]

    return sum([source_class == y_class[i[0]] for i in similarity_list]) / len(similarity_list)


if __name__ == "__main__":
    dir_name = "news_data"
    file_list = get_file_list(dir_name)
    file_list = [os.path.join(dir_name, file_name) for file_name in file_list]

    X_text, y_class = get_conetents(file_list)

    corpus = get_corpus_dict(X_text)
    print("Number of words : {0}".format(len(corpus)))
    X_vector = get_count_vector(X_text, corpus)
    source_number = 10

    result = []

    for i in range(80):
        source_number = i

        similarity_score = get_similarity_score(X_vector, source_number)
        # 가장 유사한 값 10개
        similarity_news = get_top_n_similarity_news(similarity_score, 10)
        accuracy_score = get_accuracy(similarity_news, y_class, source_number)
        result.append(accuracy_score)
    print(sum(result) / 80) # 10개를 뽑으면 7개는 야구면 야구, 축구면 축구 카테고리가 나옴...
```
