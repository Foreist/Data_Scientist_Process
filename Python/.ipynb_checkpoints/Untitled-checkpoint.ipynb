```python
from itertools import count
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

# 주소값 계속 갱신해야 함. 끝날 때까지.
html = 'http://www.mcdonalds.co.kr/www/kor/findus/district.do?pageIndex={page}&sSearch_yn=Y&skey=2&skey1=&skey2=&skeyword=%EC%84%9C%EC%9A%B8&skey4=&skey5=&skeyword2=&sflag1=&sflag2=&sflag3=&sflag4=&sflag5=&sflag6=&sflag=N'
name = []
call = []
add1 = []
add2 = []

# 계속 돌림
for pg in count():
    pgpg = pg + 1
    url = urlopen(html.format(page = pgpg))
    soup = BeautifulSoup(url, 'lxml')
    all_data = soup.find_all('dl', 'clearFix')

    # clearFix 다 찾았는데 0개이면 break
    if len(all_data) is 0:
        print("End")
        break
            
    print('현재 {}장을 크롤링 중입니다.'.format(pg))
    
    for i in all_data:
        i = i.text
        name.append(i.split('\n')[1].replace(' ', ''))
        call.append(i.split('\n')[2].replace('\r', ''))
        # 도로명만 입력되어 있는 게 있다. 그말은 add1, 2 둘 다 고쳐야 함.
        # 주소에 도로명 없으면 주소에는 주소, 도로명에는 도로명 다 받고
        if '[' not in (i.split('\n')[6]):
            add1.append(i.split('\n')[6])
            add2.append(i.split('\n')[8])
        # 아니면 주소는 받지 말고(밀려서 도로명이 들어옴)
        # 도로명밖에 없는 건 주소 자리에 도로명이 간 거니까
        else:
            add1.append(None)
            add2.append(i.split('\n')[6])
        
```
