```python
# 위키로 알아보는 select 간단요약

from bs4 import BeautifulSoup 
from urllib.request import urlopen
from urllib import parse

from IPython.display import Image 
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
```

```python
url = 'https://ko.wikipedia.org/wiki/%EC%95%88%ED%86%A0%EB%8B%88%EC%98%A4_%EB%B9%84%EB%B0%9C%EB%94%94'
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
soup.title
```

```python
# table 타입 전부 다 가져오기  (ex: <table class...)
soup.select('table')
```

```python
# 클래스명이 inforbox vcard인 거 전부 다 가져오기
soup.select('.infobox.vcard')
```

```python
# 태그 안의 태그를 보고 싶을 때
soup.select('th > div')
# 와 띄어쓰기의 차이 '  > ' = 바로 손 아래 자식, ' ' = 자손(몇 번째인지는 모름)
```

```python
soup.select('tr > td')
```

```python
soup.select('table.infobox.vcard span')
```

## >와 띄어쓰기의 차이

```python
soup.select('div a') # div 태그 하위에 a라는 게 있긴 있는데 몇 번째인지는 모르니까 다 가져와
```

```python
soup.select('div > a')# div 태그 하위에 a라는 게 있긴 있는데 내 직계 자식이야
```
