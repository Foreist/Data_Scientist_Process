# Expressing conditional logic as array operations

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
result = [(x if c else y)
          for x, y, c in zip(xarr, yarr, cond)]
result # 각 변수 인자 1개씩 빼서 True면 x 송출, 아니면 y 송출
```

```python
result = np.where(cond, xarr, yarr) # where(조건, x, y)
result
```

```python
arr = randn(4, 4) # 0 ~ 1 균일분포 난수를 4 * 4 배열로 생성
arr
```

```python
np.where(arr > 0, 2, -2)
```

```python
np.where(arr > 0, 2, arr) # false면 arr 그대로 출력
```

# Methods for boolean arrays

```python
arr = randn(100) # 가우시안 100개 생성
arr
```

```python
(arr > 0).sum() # arr 원소 중 0 초과인 것들만 다 더하긔
```

```python
bools = np.array([False, False, True, False])
```

```python
bools.any() # 이 중에 하나라도 True면 True
```

```python
bools.all() # 원소 중 전부 True면 True, 아니면 False
```

# Sorting

```python
arr = randn(8)
arr.sort()
arr # 정렬 변수값을 지정해주지 않으면 원본이 변함. sort의 기본값은 결과를 반환하지 않음
```

```python
arr = randn(5, 3); arr
```

```python
arr.sort(1); arr # column 기준으로 정렬함.
```

```python
large_arr = randn(1000)
large_arr.sort(); large_arr
```

```python
large_arr[int(0.25 * len(large_arr))] # 25%(2분위)를 보고 싶으면
```

```python
names = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
names
```

```python
np.unique(names) # == sorted(set(names))
```
