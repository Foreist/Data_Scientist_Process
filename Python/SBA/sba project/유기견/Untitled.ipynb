```python
import pandas as pd
data = pd.read_excel('sample.xlsx')
```

```python
data.head()
```

```python
from datetime import datetime
data['접수일'] = pd.to_datetime(data['접수일'], format = '%Y%m%d')
```

```python
data['나이'].value_counts()
```

```python
data['나이'] = data['나이'].astype(int)
```

```python
for i in data[data['나이'].isin([' ', '`', '-', '0', '1', '2', '3', '4', '5', '10', '20118', '208'])].index:
    data.drop(i, axis = 0, inplace = True)
```

```python
a = [i for i in data['나이']]
```

```python
year = [i.year for i in data['접수일']]
```

```python
year
```

```python
data['나이12'] = [abs(i - j) for i, j in zip(a, year)] 
```

```python
data['나이12'].value_counts()
```
