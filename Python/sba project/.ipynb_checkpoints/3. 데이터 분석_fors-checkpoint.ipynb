```python
count = 0
all_count = 0
for i in data['종'].unique():
    if  (1000 < len(data['종'][data['종'].values == i])) == True:
        print('{}'.format(i))
        count += 1
        all_count += len(data['종'][data['종'].values == i])
print(count, all_count)
```

## 상위 10%가 차지하는 개체 수

```python
# 종 unique 갯수 확인
len(data['종'].unique()) # 149개
```

```python
# 반올림하여 150개라고 하고, 10%는 15개.
# 종의 상위 15개만 출력
data['종'].value_counts()[:15]
```

```python
# 종의 상위 10%가 공교롭게도 개체수 1000개 이상부터이다.
#전체 유기견 개체 수의 비율을 얼마나 차지할까?
all_row_counts = len(data['종']); all_row_counts # 전체 유기견 수
```

```python
top10per_species_sum = sum([i for i in data['종'].value_counts()[:15]]); top10per_species_sum
```

```python
top10per_species_sum / all_row_counts * 100
# 종의 상위 10%가 전체 개체 수에서 차지하는 비중은 92.8%
# 반대로 종의 하위 90%가 전체 개체 수에서 차지하는 비중은 7%
```

```python
count = 0
all_count = 0
for i in data['종'].unique():
    if  (1000 < len(data['종'][data['종'].values == i])) == True:
        print('{}'.format(i))
        count += 1
        all_count += len(data['종'][data['종'].values == i])
print(count, all_count)
```
