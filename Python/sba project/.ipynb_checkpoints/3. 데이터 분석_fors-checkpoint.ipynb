# 종

```python
# 상위 10%가 차지하는 개체 수
# 시기별로 버려지는 특정 종이 왜 그 당시에 유기가 되었을까?
# 서연이 한글 파일에 강아지 입양 수 조사 있음
# 성견이 될 때 가장 유기가 많이 됨(1살 ~ 2살) 
count = 0
all_count = 0
for i in data['종'].unique():
    if  (1000 < len(data['종'][data['종'].values == i])) == True:
        print('{}'.format(i))
        count += 1
        all_count += len(data['종'][data['종'].values == i])
print(count, all_count)
```

```python
count = 0
all_count = 0
for i in data['종'].unique():
    if  (1000 > len(data['종'][data['종'].values == i])) == True:
        print('{}'.format(i))
        count += 1
        all_count += len(data['종'][data['종'].values == i])
print(count, all_count)
```
