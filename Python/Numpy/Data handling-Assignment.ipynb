```python
import numpy as np
```

```python
def n_size_ndarray_creation(n, dtype = np.int):
    X = np.array(np.arange(n ** 2), dtype = dtype).reshape(n, n)
    return X
```

```python
n_size_ndarray_creation(4, float)
```
