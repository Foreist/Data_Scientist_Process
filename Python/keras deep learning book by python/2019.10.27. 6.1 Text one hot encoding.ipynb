```
import numpy as np
# data
samples = ['The cat sat on the mat.', 'The dog ate my homework.']

# 데이터 모든 토큰의 인덱스를 구축
token_index = {}

for sample in samples:
  # split으로 떼어냄. 실전에서는 특수문자 다 사용함.
  print(sample)
  for word in sample.split():
    # token_index안에 word가 없을 경우(중복 허용 안 하는듯)
    if word not in token_index:
      # index 0은 관례상 사용하지 않는다고 함
      token_index[word] = len(token_index) + 1

  print(token_index)
```

```
# 샘플을 벡터로 변환
# 각 샘플당 max_length까지만 단어로 사용
max_length = 10
# len(samples) = 2, max_leng = 10, max(token_index.value()) + 1 = 10 + 1 = 11
results = np.zeros((len(samples), max_length, max(token_index.values()) + 1))
print(results)
```

```
# 문자 수준 one hot encoding

import string
characters = string.printable
characters # 모든 출력 가능한 아스키 문자
```

```
max_length = 50
# 2, 50, 101
results = np.zeros((len(samples), max_length, max(token_index.values()) + 1))
for i, sample in enumerate(samples):
  for j, character in enumerate(sample[:max_length]):
    index = token_index.get(character)
    results[i, j, index] = 1.
results
```

```
# 케라스를 사용한 단어수준의 one hot encoding
from keras.preprocessing.text import Tokenizer

# 가장 빈도가 높은 1천 개만 선택하도록
tokenizer = Tokenizer(num_words = 1000)
tokenizer.fit_on_texts(samples) # 단어 인덱스 구축

# 문자열을 정수 인덱스의 리스트로 변환
sequences = tokenizer.texts_to_sequences(samples)
sequences

# texts_to_matrix는 sample을 texts_to_sequence를 하고 sequences_to_matrix 한 것과 같다
```

```
tokenizer.sequences_to_matrix(sequences)
```

```
# 직접 one hot 이진 벡터 표현을 얻을 수 있음
one_hot_results = tokenizer.texts_to_matrix(samples, mode = 'binary')
one_hot_results.shape, one_hot_results
```

```
word_index = tokenizer.word_index
print('Found %s unique tokens' % len(word_index))
print(word_index)
```

```
# 1천 개 이상이 있다면 해싱 충돌이 늘어나고 인코딩의 정확도가 감소
dimensionality = 1000
max_length = 10

results = np.zeros((len(samples), max_length, dimensionality))
for i, sample in enumerate(samples):
  for j, word in list(enumerate(sample.split()))[:max_length]:
    # 각 단어를 해싱해서 0, 1천 사이 랜덤 정수 인덱스로 변환
    index = abs(hash(word)) % dimensionality
    results[i, j, index] = 1.
    print(i, j, index)
```
